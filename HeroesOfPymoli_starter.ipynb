```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# count number of unique user names
player_count = len(purchase_data["SN"].unique())
print("Player Count: ", player_count)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#unique items
unique_items = len(purchase_data["Item ID"].unique())

#average price
average_price = purchase_data["Price"].mean()

#count of purchases
purchase_count = len(purchase_data["Purchase ID"].unique())

#total revenue
total_revenue = purchase_data["Price"].sum()

#purchasing summary table
purchasing_total = pd.DataFrame({" Number of Unique Items": unique_items,
                              "Average Purchase Price": [average_price],
                              "Total Number of Purchases": [purchase_count],
                              "Total Revenue": [total_revenue]})
purchasing_total.round(2)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#group by gender
gender_count = purchase_data.groupby("Gender")["SN"].nunique()
gender_percentage = gender_count/player_count

#gender summary table 
gender_summary = pd.DataFrame({"Gender Count": gender_count, 
                               "Gender Percentage": gender_percentage})

gender_summary.round(2)
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchases by gender
gender_purchase_count = purchase_data.groupby("Gender")["Item Name"].count()

#average price by gender
gender_average = purchase_data.groupby("Gender")["Price"].mean()

#total price by gender
gender_total = purchase_data.groupby("Gender")["Price"].sum()

#average purchase total per person by gender
average_gender_total = gender_total/gender_count

#gender analysis summary

gender_analysis = pd.DataFrame({"Purchase Count": gender_purchase_count,
                   " Average Purchase Price": gender_average,
                   "Total Purchase Value": gender_total,
                   "Average Purchase Total": average_gender_total})
gender_analysis.round(2)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#age bins 
bins = [0, 10, 15, 20, 25, 30, 35, 40]

#age ranges
age_ranges = ["<10", "10-14", "15-19", "20-24", "30-34", "35-39", "40+"]

#cut and add ages into bins
pd.cut(purchase_data["Age"], bins, labels=age_ranges)

#add age range column
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=age_ranges)

#age percentages
age_percentage = purchase_data["Age Range"].value_counts()/purchase_count
age_percentage

#total purchases by age
purchase_age = purchase_data.groupby("Age Range")["Item Name"].count()

#age summary
age_summary = pd.DataFrame({"Total Count": purchase_age,
                            "Percentage of Players": age_percentage})

age_summary.round(2)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#average price by age
age_average = purchase_data.groupby("Age Range")["Price"].mean()

#total price by age
age_total = purchase_data.groupby("Age Range")["Price"].sum()

#average purchase total by age
average_per_age = age_total/player_count

#age analysis
age_analysis = pd.DataFrame({"Purchase Count": purchase_age,
                             "Average Purchase Price": age_average,
                             "Total Purchase Value": age_total,
                             "Average Purchase Total by Age":average_per_age})

age_analysis.round(2)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#player purchases
player_purchase_count = purchase_data.groupby("SN")["Item Name"].count()

#average price per player 
player_average = purchase_data.groupby("SN")["Price"].mean()

#total spend per player
player_spend = purchase_data.groupby("SN")["Price"].sum()

#total player df
total_player_df = pd.DataFrame({"Purchase Count": player_purchase_count,
                               "Average Purchase Price": player_average,
                               "Total Purchase Value": player_spend})

#top spenders
top_5  = total_player_df.sort_values("Total Purchase Value", ascending=False)

top_5.head().round(2)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#purchased items count
items_purchase_count = purchase_data.groupby(['Item ID', 'Item Name']).count()['Price'].rename('Purchase Count')

#average price of item
items_average_price = purchase_data.groupby(['Item ID', 'Item Name']).mean()['Price'].rename('Average Price')

#total items price
items_purchase_count = purchase_data.groupby(['Item ID', 'Item Name']).sum()['Price'].rename('Total Purchase Value')


# items df
purchased_items = pd.DataFrame({'Purchase Count': items_purchase_count,
                               'Item Price': items_average_price,
                               'Total Purchase Value': items_purchase_count})

#most popular 
most_popular_items = purchased_items.sort_values('Purchase Count', ascending=False)

#top 5 popular
most_popular_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable = purchased_items.sort_values('Total Purchase Value', ascending=False)
most_profitable.head()
```
